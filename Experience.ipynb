```python
data = pd.read_json('talents_sample.json')
```

```python

# Make a empty dataframe 
experience = pd.DataFrame()
for i in range(len(data)):
    
    #Making a dataframe of data['Experience'][i] column 
    t =  pd.DataFrame(data['Experience'][i])
    
    #If company name is not there than remove that row
    for j in range(len(t)):
            if t['companyName'][j] == None:
                t=t.drop(j)            
    t=t.reset_index(drop=True)
    
    # dataframe is null
    if len(t) ==0:
        ids = data['_id'][i]
        Most_Worked_Company_Name = "Na"
        Least_Worked_Company_Name = "Na"
        Total_experience=0
        Career_gap=0
        data_info1 = {'Ids':ids,'Most_Worked_Company_Name':Most_Worked_Company_Name,'Least_Worked_Company_Name':Least_Worked_Company_Name,'Total_experience':Total_experience,'Career_gap':Career_gap}
        experience = experience.append(data_info1,ignore_index=True).reset_index(drop=True)
        continue
    
    
    
    #replace current or Present with todays date
    t = t.replace(to_replace =['current','Present'], value =date.today())
    
    # conerting start_date and end_date string column to date fromate
    t['startDate'] = pd.to_datetime(t['startDate'], utc=True)
    t['endDate'] = pd.to_datetime(t['endDate'], utc=True)
    
    # sort the dateframe with respect to startdate and reset index
    t=t.sort_values(by=['startDate'], ascending=True)
    t=t.reset_index(drop=True)
    
    # for dealing the case in which a person is working in the same company and its position change 
    # But its experience is = total time spend in that company
    
    for j in range(1,len(t)):
        if j >= len(t):
            break
        if t['companyName'][j] == t['companyName'][j-1] and t['endDate'][j-1] == t['startDate'][j]:
            t['endDate'][j-1] = t['endDate'][j]
            t=t.drop(j)
            t=t.reset_index(drop=True)
    
    # Condition for removing row don't having both StartDate and EndDate not present
    for j in range(len(t)):
        if j < len(t):
              if type(t['endDate'][j]) ==pd._libs.tslibs.nattype.NaTType and type(t['startDate'][j]) ==pd._libs.tslibs.nattype.NaTType:
                    t=t.drop(j)
                    t=t.reset_index(drop=True)
    
    # if 'employmentDuration' column  is present and endate is not than we add employee duration in startDate 
    if 'employmentDuration' in t:
       for j in range(len(t)):
            if type(t['endDate'][j]) ==pd._libs.tslibs.nattype.NaTType:
                t['endDate'][j] = t['startDate'][j]+pd.offsets.DateOffset(years=t['employmentDuration'][j]['years'], months=t['employmentDuration'][j]['months'])
                t['endDate'][j]=pd.to_datetime(t['endDate'][j], utc=True)
    
    # last row of dataframe dont have enddate than make it todays date            
    if type(t['endDate'][len(t)-1])==pd._libs.tslibs.nattype.NaTType:
                t['endDate'][len(t)-1] = pd.to_datetime(date.today(), utc=True)
                           
    # if other than last row,  enddate is not present than make endate equal  start date of next row   
    for j in range(len(t)-1):
        if type(t['endDate'][j]) ==pd._libs.tslibs.nattype.NaTType:
            t['endDate'][j] = t['startDate'][j+1]
    
    # for handling duplicate value with consideration that, a person can work only at 1 company in an duration
    # if start date of any 2 column is same than lowest enddate value row get deleted
    for j in range(1,len(t)):
        if j >= len(t):
            break

        if t['startDate'][j] == t['startDate'][j-1]:
              if t['endDate'][j] > t['endDate'][j-1]:
                  t=t.drop(j-1)
                  t=t.reset_index(drop=True)
              else:
                  t=t.drop(j)
                  j=j-1
                  t=t.reset_index(drop=True)    
    
    # Extracting id of 't' dataframe from original data 
    ids = data['_id'][i]
    
    # for get total experience of a particular person 
    # first we make a sperate column in which equals to t['endDate']-t['startDate'] and then convert it to years
    t['exp'] =(t['endDate']-t['startDate'])/np.timedelta64(1, 'Y')
    # then take the total sum of that column as "Total_experience"
    Total_experience= float("{:.2f}".format(t['exp'].sum()))
   
    # for finding most worked company we get the index value of row which have highest exp than make than index company name as Most_Worked_Company_Name 
    Most_Worked_Company_Name=t['companyName'][t['exp'].idxmax(axis = 0)]
    # for finding Least worked company we get the index value of row which have minimum exp than make than index company name as Least_Worked_Company_Name
    Least_Worked_Company_Name=t['companyName'][t['exp'].idxmin(axis = 0)]
    
    # for getting extracting career gap 
    Career_gap= (float("{:.2f}".format((pd.to_datetime(date.today(), utc=True)-pd.to_datetime(t['startDate'][0]))/np.timedelta64(1, 'Y'))))-Total_experience
    
    # now adding the information we exteract of particular row in experience datframe 
    data_info1 = {'Ids':ids,'Most_Worked_Company_Name':Most_Worked_Company_Name,'Least_Worked_Company_Name':Least_Worked_Company_Name,'Total_experience':Total_experience,'Career_gap':Career_gap}
    experience = experience.append(data_info1,ignore_index=True).reset_index(drop=True)
    
```
